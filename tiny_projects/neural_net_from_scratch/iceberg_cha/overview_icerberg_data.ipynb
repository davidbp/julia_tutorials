```julia
n = length(data[1]["band_1"])
costat = Int(sqrt(n))
```

```julia
data[1]["is_iceberg"], data[3]["is_iceberg"]
```

```julia
# Non iceberg sample

x1 = data[1]["band_1"];
x1 = Array{Float32}(x1);
xim1 = reshape(x1, costat, costat);

x2 = data[1]["band_2"];
x2 = Array{Float32}(x2);
xim2 = reshape(x2, costat, costat);

figure, axis = PyPlot.subplots(1,2);

axis[1][:imshow](xim1)
axis[1][:set_title]("band 1")

axis[2][:imshow](xim2)
axis[2][:set_title]("band 2")

PyPlot.show()
```

```julia
## ICERBERG sample

x1 = data[3]["band_1"];
x1 = Array{Float32}(x1);
xim1 = reshape(x1, costat, costat);

x2 = data[3]["band_2"];
x2 = Array{Float32}(x2);
xim2 = reshape(x2, costat, costat);

figure, axis = PyPlot.subplots(1,2);

axis[1][:imshow](xim1)
axis[1][:set_title]("band 1")

axis[2][:imshow](xim2)
axis[2][:set_title]("band 2")

PyPlot.show()
```
